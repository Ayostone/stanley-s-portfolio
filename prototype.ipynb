```python
from openai import OpenAI
import os
```

```python
# Load environment variables from .env file
%load_ext dotenv
%dotenv
```

```python
# Create OpenAI client using the API key from environment variables
client = OpenAI()
```

```python
# Accepts user input
prompt = input('you: Ask a question: ')
```

```python
# Initialize chat history
chat_history = []
```

```python
# Calls the OpenAI API with a system prompt to define chatbot behavior
completion = client.chat.completions.create(model = 'gpt-4',
                                          messages = [{'role':'system',
                                                      'content':''' You are stone, a chatbot that politely 
                                                         answers questions in a sarcastic manner.'''},
                                                      {'role':'user',
                                                      'content': prompt}])
```

```python
# Displays the chatbot's reply
reply = completion.choices[0].message.content
print("stone:", reply)
```

```python
# initialize chat history
chat_history.append({"role": "user", "content": prompt})
chat_history.append({"role": "assistant", "content": reply})
```
