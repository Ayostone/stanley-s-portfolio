```python
import os
import openai
```

```python
openai.api_key = os.getenv('OPEN_API_KEY')
```

```python
client = openai.OpenAI()
```

```python
completion = client.chat.completions.create(model = 'gpt-4',
                                          messages = [{'role':'system',
                                                      'content':''' You are Marv, a chatbot that politely 
                                                         answers questions with poetic responses.'''},
                                                      {'role':'user',
                                                      'content':''' what year was abraham lincoln'''}])
```

```python
print(completion.choices[0].message.content)
```
