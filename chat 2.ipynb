```python
from openai import OpenAI
import os
```

```python
# Load environment variables from .env file
%load_ext dotenv
%dotenv
```

```python
def load_api_key():
    if 'OPENAI_API_KEY' not in os.environ:
        raise ValueError("OPENAI_API_KEY not found in environment variables.")
```

```python
# Create OpenAI client using the API key from environment variables
client = OpenAI()
```

```python
# Accepts user input
prompt = input('you: Ask a question: ')
```

```python
# Initialize chat history
chat_history = []
```

```python
# Constructs the message list for the chat completion request
def chat_messages(prompt, chat_history, system_prompt):
    system_message = system_prompt if system_prompt is not None else ''' You are stone, a chatbot that sarcastically answers questions and provides unhelpful information.'''
    return[{'role':'system',
            'content': system_message},
            *chat_history, 
            {'role':'user', 'content': prompt}]
```

```python
# Sends the chat completion request to the OpenAI API
def chat_request(messages):
    completion = client.chat.completions.create(model = 'gpt-4',
                                              messages=messages)
    return completion.choices[0].message.content
```

```python
# Main chat function
def chat(prompt, chat_history, system_prompt):
    messages = chat_messages(prompt, chat_history, system_prompt)
    
    reply = chat_request(messages)
    return reply
    # update chat history
    chat_history.append({"role": "user", "content": prompt})
    chat_history.append({"role": "assistant", "content": reply})

```
